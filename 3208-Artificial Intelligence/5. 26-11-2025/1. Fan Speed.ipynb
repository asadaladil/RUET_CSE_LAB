```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input Define:
n=1+1
temperature={
    'cold':[20,50],
    'warm':[40,75],
    'hot':[65,100]
}

# Output Define:
fan_speed={
    'slow':[15,40],
    'medium':[35,50],
    'fast':[45,60]
}

y=[0,1,0]
plt.figure(figsize=(10,8))
plt.subplot(n,1,1)
plt.xticks(np.arange(10,105,5))
plt.grid(True)
plt.title('Triangular Membership Function')
plt.xlabel('Temperarture')
plt.ylabel('Membership Degree')
for i,j in temperature.items():
    x=[]
    for k in j:
        x.append(k)
    x.insert(1,np.sum(j)/2)
    plt.plot(x,y,label=i)
    plt.legend()

plt.subplot(n,1,2)
plt.xticks(np.arange(10,75,5))
plt.grid(True)
plt.xlabel('Fan Speed')
plt.ylabel('Membership Degree')
for i,j in fan_speed.items():
    x=[]
    for k in j:
        x.append(k)
    x.insert(1,np.sum(j)/2)
    plt.plot(x,y,label=i)
    plt.legend()
    
plt.show()
```

```python
def rules(temp):
    if(temp=='hot'):
        return np.sum(fan_speed['fast'])/2
    elif(temp=='warm'):
        return np.sum(fan_speed['medium'])/2
    elif(temp=='cold'):
        return np.sum(fan_speed['slow'])/2
    
    
```

```python
def membership_degree(x,s):
    a=temperature[s][0] # minimum value
    b=np.mean(temperature[s]) # mid value
    c=temperature[s][1] #max value
    if x>=c:
        return 0
    if x>b:
        return (c-x)/(c-b)
    if x>a:
        return (x-a)/(b-a)
    else:
        return 0
```

```python
def Fuzzy(x):
    membership_value=[]
    for i,j in temperature.items():
        result=membership_degree(x,i) # take the membership value from given input
        membership_value.append(result)

    print(membership_value)

    crisp_speed=[]
    for i in temperature:
        crisp_speed.append(rules(i)) # storing the rule
    
    print(crisp_speed)
        
    crisp_speed=np.array(crisp_speed)
    try:
        crisp_speed=crisp_speed*np.array(membership_value) #multiplication
        crisp_speed=np.sum(crisp_speed)/np.sum(membership_value) # Divide
    except:
        crisp_speed=0
        

    return crisp_speed
```

```python
x=45
print(f"{Fuzzy(x):.3f}")
```
