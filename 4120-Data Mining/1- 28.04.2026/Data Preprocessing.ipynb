```python
import pandas as pd
import numpy as np

df = pd.read_csv("Titanic-Dataset.csv")

print("Initial shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
```

```python
# print(df['Cabin'].unique()[67])

print(df['Ticket'].unique())
```

```python
df_drop = df.dropna()

# Option 2: Fill missing values
df_fill = df.fillna({
 'Age': df['Age'].mean(),
 'Cabin': df['Cabin'].unique()[df['Cabin'].unique().argmax()],
 'Embarked':df['Embarked'].unique()[df['Embarked'].unique().argmax()]
})
```

```python
print("\nMissing values:\n", df_fill.isnull().sum())
```

```python
# One-hot encoding using pandas
df_encoded = pd.get_dummies(df_fill, columns=['Gender', 'Embarked'],
drop_first=True)

print(df_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

num_cols = ['Age', 'Fare']

# Standardization: mean = 0, std = 1
scaler_std = StandardScaler()
df_std = df_encoded.copy()
df_std[num_cols] = scaler_std.fit_transform(df_std[num_cols])
print(df_std)

```

```python
# Normalization: range [0, 1]
scaler_minmax = MinMaxScaler()
df_norm = df_encoded.copy()
df_norm[num_cols] = scaler_minmax.fit_transform(df_norm[num_cols])
print(df_norm)
```

```python
before = df_norm.shape[0]
df_clean = df_norm.drop_duplicates()
after = df_clean.shape[0]
print(f"Removed {before - after} duplicate rows")
```
