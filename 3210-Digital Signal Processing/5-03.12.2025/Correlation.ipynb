```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def correlation(x,x0,y,y0):
    t=[]
    rxy=[]
    mxx=-100000
    # rxy(n)= sum(k=-inf to inf) x(k)*y(k-n)
    mx=max(len(x),len(y))
    for n in range(-mx,mx+1):
        rxy.append(0)
        t.append(n)
        for k in range(-mx,mx+1):
            try:                       # if positive index exist?
                if k>=(-x0): a=x[k+x0] # if negative index exists?
                else: a=0
            except: a=0
            
            try:    
                if (k-n)>=(-y0): b=y[k+y0-n]
                else: b=0
            except: b=0
            
            rxy[-1]+=a*b
        
        if mxx<=rxy[-1]:
            delay=t[-1]
            mxx=rxy[-1]
        
    return t,rxy,(delay,mxx)
```

```python
x1=[1,2,4,7,12]
x2=[0,1,1,2,6,9,14]

t,rxy,result=correlation(x1,0,x2,0)
print("Time Sequence:",t)
print("Correlation Values:",rxy)
print("Delay:",result[0])
print("Maximum Value:",result[1])
```

```python
plt.stem(t,rxy)
plt.grid(True)
plt.show()
```

```python
# Cross Correlation
f=1
t=np.linspace(-1,1,100)
sinx=np.sin(2*np.pi*f*t)

shifted_sign=list(sinx)
t=5
for i in range(t):
    shifted_sign.insert(0,0)

shifted_sign=np.array(shifted_sign)

cross_corr=np.correlate(sinx,shifted_sign,'full')

plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.stem(sinx)
plt.grid(True)

plt.subplot(3,1,2)
plt.stem(shifted_sign)
plt.grid(True)

plt.subplot(3,1,3)
plt.stem(cross_corr)
plt.grid(True)

plt.show()
```

```python
# Auto correlation
f=1
t=np.linspace(-1,1,100)
cosx=np.cos(2*np.pi*f*t)
corr=np.correlate(cosx,cosx,'full')

plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.stem(cosx)
plt.grid(True)

plt.subplot(2,1,2)
plt.stem(corr)
plt.grid(True)
plt.show()
```
