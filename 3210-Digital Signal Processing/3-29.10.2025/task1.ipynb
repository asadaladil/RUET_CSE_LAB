```python
import numpy as np
import matplotlib.pyplot as plt

x_n=np.array([2,0,2])
# x_n=np.array([1,2,4,8])
x_t=0
h_n=[1,1,1]
h_t=1
length=len(x_n)+len(h_n)-1
for i in range(length-len(h_n)+1):
    h_n.append(0)
h_n=np.array(h_n)
print(length,h_n)

```

```python
# i) 2del(n)

t=np.array([ i-x_t for i in range(len(x_n))])
del_n=np.array([0 for i in range(len(x_n))])
print(del_n)
del_t=x_t
for i in t:
    if i==0:
        del_n[i+x_t]=x_n[i]
    else:
        del_n[i]=0

print(del_n) 
del_n=np.array(del_n)
print(del_n,t)

plt.grid(True)
plt.stem(t,del_n)
plt.show()

```

```python
# ii) 2h(n)

h1=h_n*del_n[0]
print(h1)

t=np.array([ i-h_t for i in range(len(h_n))])
plt.grid(True)
plt.stem(t,h1)
plt.show()
```

```python
# iii) 2del(n-2)

t=np.array([ i-x_t for i in range(len(x_n))])
del_n=np.array([0 for i in range(len(x_n))])
print(del_n)
del_t=x_t
for i in t:
    if i==2:
        del_n[i+x_t]=x_n[i]
    else:
        del_n[i]=0

print(del_n) 
del_n=np.array(del_n)
print(del_n,t)

plt.grid(True)
plt.stem(t,del_n)
plt.show()
```

```python
# iv) 2h(n-2)
for i in range(2):
    h_n=h_n[:-1]
    h_n=np.insert(h_n,0,0)
h2=h_n*del_n[2]
print(h2)

t=np.array([ i-h_t for i in range(len(h_n))])
plt.grid(True)
plt.grid(True)
plt.stem(t,h2)
plt.show()
```

```python
# v) h(n)+h(n-2)

y_n=h1+h2
print(y_n)

t=np.array([ i-h_t for i in range(len(h_n))])
plt.grid(True)
plt.stem(t,y_n)
plt.show()
```
