```python
import numpy as np
import matplotlib.pyplot as plt

f=1
n=np.linspace(-1,1,1000)
x_n=np.sin(2*np.pi*n*f)

plt.plot(x_n)
plt.show()
```

```python

noise=np.random.rand(len(n))
# print(noise)

noisy=x_n+noise
# print(noisy)

plt.plot(noisy)
plt.show()
```

```python
filtered=np.ones(100)/100

after_filter=np.convolve(noisy,filtered)
plt.plot(after_filter)
plt.show()
```
