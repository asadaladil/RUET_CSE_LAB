```python
import numpy as np
import matplotlib.pyplot as plt

x_n=np.array([1,2,3])
# x_n=np.array([1,2,4,8])
lenx=len(x_n)
x_t=0
h_n=np.array([4,5,6])
# h_n=[1,1,1,1,1,1,1]
h_t=1
y_t=h_t
length=len(x_n)+len(h_n)-1
h_n=h_n[::-1]

for i in range(len(h_n)):
    x_n=np.insert(x_n,0,0)

for i in range(len(h_n)):
    x_n=np.append(x_n,0)

l=length-len(h_n)-h_t+1
for i in range(l):
    h_n=np.append(h_n,0)
    h_t-=1

for i in range(len(x_n)-len(h_n)):
    h_n=np.append(h_n,0)


print(h_n,x_n,h_t)
```
