```python
# Signal Manipulation
import numpy as np
import matplotlib.pyplot as plt

n=np.array([i for i in range(-8,10,1)])
signal=[]

def value(s):
    if s>=0 and s<=6 and round(s)==s:
        return 1
    return 0

signal=np.array(list(map(value,n)))

# Original Signal
plt.stem(n,signal)
plt.xticks(n)
plt.grid(True)
plt.show()
```

```python
# Shifting:

k=2
modulated=n+k
plt.stem(modulated,signal)
plt.xticks(modulated)
plt.grid(True)
plt.show()
```

```python
# Reversal

modulated=-modulated
print(modulated)
plt.stem(modulated,signal)
plt.xticks(modulated)
plt.grid(True)
plt.show()
```

```python
# Amplitude Scaling:

A=3
signal=A*signal
plt.stem(modulated,signal)
plt.xticks(modulated)
plt.grid(True)
plt.show()
```

```python
# Time Scaling:

T=2

def new_signal(x,y):
    for i in range(len(x)):
        if round(x[i])!=x[i]:
            y[i]=0
    return y
modulated=modulated/T
signal=new_signal(modulated,signal)
plt.figure(figsize=(10,5))
plt.stem(modulated,signal)
plt.xticks(modulated)
plt.grid(True)
plt.show()
```
