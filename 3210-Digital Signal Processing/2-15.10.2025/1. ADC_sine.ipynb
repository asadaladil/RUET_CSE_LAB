```python

import numpy as np
import matplotlib.pyplot as plt

min=-1
max=1
t=np.linspace(min,max,1000)
f=4
x_t=np.cos(2*np.pi*f*t)

plt.figure(figsize=(10,4))
plt.plot(t,x_t)
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("Plot of cosine function")
plt.grid(True)
plt.show()
```

```python
#sampling
Fs=30 # Fs>F
Ts=1/Fs
sample_time=np.arange(min,max,Ts)

xt_sample=np.cos(2*np.pi*f*sample_time)
plt.stem(sample_time,xt_sample)
plt.plot(t,x_t,'r')
plt.grid(True)
plt.show()
```

```python
xmin=np.min(xt_sample)
xmax=np.max(xt_sample)

level=11
quantization_step=(xmax-xmin)/(level-1)

quantized_signal=[]

for i in xt_sample:
    value=round(i/quantization_step)*quantization_step
    quantized_signal.append(value)
quantized_signal=np.array(quantized_signal)


plt.figure()
plt.stem(sample_time,quantized_signal)

for label_value in np.arange(xmin, xmax + quantization_step, quantization_step):
    plt.axhline(y=label_value, color='g', linestyle='--')

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Quantized Signal with Quantization Labels")
plt.show()
```
