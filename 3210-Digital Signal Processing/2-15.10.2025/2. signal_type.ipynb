```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Unit sample signal

n=np.array([i for i in range(-5,6,1)])
del_n=[]
for i in n:
    if i==0:
        del_n.append(1)
    else:
        del_n.append(0)
del_n=np.array(del_n)
plt.xticks(n)
plt.grid(True)
plt.stem(n,del_n)
plt.show()
```

```python
# unit step signal

u=[]
for i in n:
    if i>=0:
        u.append(1)
    else:
        u.append(0)
del_n=np.array(del_n)
plt.grid(True)
plt.stem(n,u)
plt.xticks(n)
plt.show()
```

```python
# Unit Ramp signal

ur=[]
for i in n:
    if i>=0:
        ur.append(i)
    else:
        ur.append(0)
del_n=np.array(ur)
plt.grid(True)
plt.stem(n,ur)
plt.xticks(n)
plt.show()
```
