```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def correlation(x,x0,y,y0):
    t=[]
    rxy=[]
    # rxy(n)= sum(k=-inf to inf) x(k)*y(k-n)
    mx=max(len(x),len(y))
    for n in range(-mx,mx+1):
        rxy.append(0)
        t.append(n)
        for k in range(-mx,mx+1):
            try:                       # if positive index exist?
                if k>=(-x0): a=x[k+x0] # if negative index exists?
                else: a=0
            except: a=0
            
            try:    
                if (k-n)>=(-y0): b=y[k+y0-n]
                else: b=0
            except: b=0
            
            rxy[-1]+=a*b
        
    return t,rxy
```

```python
# Auto-correlation

x=np.array([1,3,5,7])
x0=0

t,rxy=correlation(x,x0,x,x0)
        
print(rxy)
print(t)
plt.stem(t,rxy)
plt.grid(True)
plt.show()
 
```

```python
# cross-corelation:

# x=np.array([2,3,5,7])
x=np.array([1,2,3,4])
x0=2
# y=np.array([1,1,3,7,9,13,2])
y=np.array([2,4,5,8,4,3])
y0=1
t,rxy=correlation(x,x0,y,y0)
print(rxy)
print(t)
plt.stem(t,rxy)
plt.grid(True)
plt.show()
 
```

```python
f=1
t=np.linspace(-1,1,1000)
sinx=np.sin(2*np.pi*f*t)
cosx=np.cos(2*np.pi*f*t)

plt.plot(sinx)
plt.plot(cosx)
plt.grid(True)
plt.legend(('sin(x)','cos(x)'))
plt.title("Periodic Signal")
plt.show()
```

```python
auto_correlate=np.correlate(sinx,sinx,mode='full')

plt.plot(auto_correlate)
plt.title("Auto-correlation of Sine Wave")
plt.grid(True)
plt.show()
```

```python
auto_correlate=np.correlate(cosx,cosx,mode='full')

plt.plot(auto_correlate)
plt.title("Auto-correlation of Cosine Wave")
plt.grid(True)
plt.show()
```
