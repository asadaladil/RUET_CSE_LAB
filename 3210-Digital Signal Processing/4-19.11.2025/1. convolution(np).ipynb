```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_n=np.array([4,8,9,10])
h_n=np.array([1,1,1,1])

y_n=np.convolve(x_n,h_n,'full')
print(y_n)
length=len(y_n)
t=[i for i in range(length)]

plt.grid(True)
plt.stem(t,y_n)
plt.show()
```

```python
y_n=np.convolve(x_n,h_n,'same')
print(y_n)
length=len(y_n)
t=[i for i in range(length)]

plt.grid(True)
plt.stem(t,y_n)
plt.show()
```

```python
y_n=np.convolve(x_n,h_n,'valid')
print(y_n)
length=len(y_n)
t=[i for i in range(length)]

plt.grid(True)
plt.stem(t,y_n)
plt.show()
```

```python
# x_n=np.array([2,0,2])
x_n=np.array([1,2,3,4])
x0=2

h_n=np.array([1,1,1])
# h_n=np.array([0.2,0.5,0.2])
h0=1

length=len(x_n)+len(h_n)-1

mx=max(len(x_n),len(h_n))

t=[]
y_n=[]
for n in range(-mx,length+1):
    y_n.append(0)
    t.append(n)
    for k in range(-mx,length+1):
        try:                         # if positive index out of range?
            if k>=(-x0): a=x_n[k+x0] # if negative index value exist?
            else: a=0
        except:a=0
        
        try:
            if (-k+n)>=(-h0): b=h_n[-k+h0+n]
            else:b=0
        except: b=0

        y_n[-1]+=a*b

print(y_n,t)

plt.stem(t,y_n)
plt.grid(True)
plt.show()
```
