```python
x_train=[]
y_train=[]
with open(f"./Test_Digit/test.txt",mode="r") as file:
    a=[]
    file=file.readlines()
    for i in file:
        if len(i.split(' '))==1: #for taking y values
            x_train.append(a)
            y_train.append(int(i))
            a=[]
            continue
        temp=[]
        for j in i.split(' '):
            try:
                temp.append(int(j))
            except:
                continue
        a.append(temp)
    
x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
import time
from Pytorch_Prediction import predict_from_array
y_pred=[]

with open("CNN_predict_custom.txt",mode="w") as file:  
    a = time.time()  
    for i in x_train:
        # file.write(str(k)+'\n')
        y_pred.append(predict_from_array(i))
b = time.time()

total_inference_time = b - a
print(f"Total Inference Time: {np.round(total_inference_time, 2)} seconds")
print(f"Average Latency per Image: {(total_inference_time / len(x_train)) * 1000:.2f} ms")
y_pred=np.array(y_pred)

correct=0

for i in range(10700):
    if y_pred[i]==y_train[i]:
        correct+=1

print((correct*100)/10700)
```
